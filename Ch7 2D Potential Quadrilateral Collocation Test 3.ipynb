## Solving the system

```python
# Compute the source terms
# The exact solution is u = x^2 
# s = 2
S = np.ones((mg.nodal2D.Nx+1, mg.nodal2D.Ny+1)) * 2
```

```python
# Initializing the solution matrix
U = np.zeros((mg.nodal2D.Nx+1, mg.nodal2D.Ny+1))
# Compute the boundary values
# The exact solution is u = x^2+y^2
# The following can be vectorized, not done yet.

for j in [0, mg.nodal2D.Ny]:
    for i in range(mg.nodal2D.Nx+1):
        x = mg.nodes_phy_x[i,j]
        y = mg.nodes_phy_y[i,j]
        U[i,j] = x**2 
        
        
for i in [0, mg.nodal2D.Nx]:
    for j in range(mg.nodal2D.Ny+1):
        x = mg.nodes_phy_x[i,j]
        y = mg.nodes_phy_y[i,j]
        U[i,j] = x**2 
# Visualization
fig, ax = plt.subplots(nrows=1, ncols=1, subplot_kw={"projection": "3d"})
ax.plot_surface(mg.nodes_phy_x, mg.nodes_phy_y, U, cmap=cm.coolwarm)
ax.set_title("BC")        
```

```python
# Set BC, alternative method
U = mg.nodes_phy_x**2 
U[1:mg.nodal2D.Nx, 1:mg.nodal2D.Ny] = 0.0
# Visualization
fig, ax = plt.subplots(nrows=1, ncols=1, subplot_kw={"projection": "3d"})
ax.plot_surface(mg.nodes_phy_x, mg.nodes_phy_y, U, cmap=cm.coolwarm)
ax.set_title("BC")
```

```python
# Preparing coefficients
coeff_xi = (mg.X_xi**2 + mg.Y_xi**2) / mg.J
coeff_eta = (mg.X_eta**2 + mg.Y_eta**2) / mg.J
coeff_mixed = (mg.X_xi * mg.X_eta + mg.Y_xi * mg.Y_eta) / mg.J
# Constructing RHS
RHS = cal_RHS(D_xi=mg.nodal2D.Dx,
              D_eta=mg.nodal2D.Dy,
              J=mg.J,
              coeff_xi=coeff_xi,
              coeff_eta=coeff_eta,
              coeff_mixed=coeff_mixed,
              U=U,
              S=S,
              N=mg.nodal2D.Nx,
              M=mg.nodal2D.Ny)
# Constructing the extended system: Cx = d
C, d = MatEqnConverter_NonsquareCollocation_Dirichlet(D_xi=mg.nodal2D.Dx,
                                                      D_eta=mg.nodal2D.Dy,
                                                      J=mg.J,
                                                      coeff_xi=coeff_xi,
                                                      coeff_eta=coeff_eta,
                                                      coeff_mixed=coeff_mixed,
                                                      RHS=RHS,
                                                      N=mg.nodal2D.Nx,
                                                      M=mg.nodal2D.Ny)
```

```python
# Solving the system
U[1:mg.nodal2D.Nx, 1:mg.nodal2D.Ny] = MatEqnSolver_NonsquareCollocation(C=C, 
                                                                        d=d, 
                                                                        N=mg.nodal2D.Nx,
                                                                        M=mg.nodal2D.Ny)
```

```python
# Comparing with the exact solution
# The exact solution is u = x^2 + y^2
U_th = mg.nodes_phy_x**2 
print("Error:", np.max(np.abs(U-U_th)))
```
