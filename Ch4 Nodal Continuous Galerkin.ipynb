```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Reloading the module
import importlib

import OrthogonalPolynomials
importlib.reload(OrthogonalPolynomials)
from OrthogonalPolynomials import *

import GeneralFunctionality
importlib.reload(GeneralFunctionality)
from GeneralFunctionality import *

from scipy.special import eval_legendre
```

```python
# Node indices: 0,1,...,N
N = 20
# Legendre Gauss Lobatto nodes
x = LegendreGaussLobattoNodes(N)
# Legendre Gauss Lobatto weights
w_LGL = LegendreGaussLobattoWeights(x)
# Barycentric weights
w_bary = BarycentricWeights(x)
# Differentiation matrix
D = PolynomialDiffMatrix(x, w_bary)
```

```python
def MatrixG(w, D):
    # Calculating the matrix \hat{G}
    # w: Legendre Gauss Lobatto weights
    # D: Differentiation matrix
    
    N = len(w)-1
    
    # Initialization
    G = np.zeros((N+1, N+1))
    
    # Calculating the entries
    for m in range(0,N+1): 
        for n in range(0,N+1): 
            for k in range(0,N+1): 
                G[m,n] += w[k] * D[k,m] * D[k,n]
            G[m,n] /= w[m]
    
    return G

G = MatrixG(w_LGL, D)
```

```python
# IC
def InitCond_1D_Diffusion(x):
    # Calculating the initial value at x
    return 1-x**2

u0 = np.zeros(N+1)
# The initial condition function may not be vectorized
# So we use a for loop to calculate the initial nodal values separately
for j in range(N+1):
    u0[j] = InitCond_1D_Diffusion(x[j])
    
# BC:
u_left = 0
u_right = 0
```

```python
def RHS_Dirichlet(t, u, N, G):
    # Evaluate the RHS of the ODE system for the 1D diffusion equation
    # u: uj: j=1,...,N-1, inner solution values
    # G: the \hat{G} in the book
    
    return -np.dot(G[1:N,1:N], u)
```

```python
# Scipy calculation
t_span = [0, 1]
t_eval=[0,0.5,1]
sol = solve_ivp(RHS_Dirichlet, 
                t_span=t_span, 
                y0=u0[1:-1], 
                method='RK45', 
                t_eval=t_eval, 
                args=(N, G))

fig = plt.figure()
ax = fig.add_subplot(111)

Nt = len(t_eval)
for it in range(Nt):
    u = np.concatenate(( np.array([u_left]), 
                         sol.y[:,it], 
                         np.array([u_right]) ))
    ax.plot(x, u, label='t={}'.format(t_eval[it]))


ax.legend()
ax.set_title(r'$u_t=u_{xx}$')
```

```python
# Interpolating to a finer grid for plotting purpose
y = np.linspace(-1,1,51)
T = InterpolatingMatrix(x, y, w_bary)

fig = plt.figure()
ax = fig.add_subplot(111)

Nt = len(t_eval)
for it in range(Nt):
    u = np.concatenate(( np.array([u_left]), sol.y[:,it], np.array([u_right]) ))
    u_newgrid = np.dot(T, u)
    ax.plot(y, u_newgrid, label='t={}'.format(t_eval[it]))


ax.legend()
ax.set_title(r'$u_t=u_{xx}$')
```

## Neumann boundary conditions

```python
u_left = 0
ux_right = 1

def RHS_Neumann(t, u, N, G, w_LGL, ux_right):
    # Evaluate the RHS of the ODE system for the 1D diffusion equation
    # u: uj: j=1,...,N, inner solution values
    # G: the \hat{G} in the book
    
    rhs = -np.dot(G[1:, 1:], u)
    rhs[-1] += ux_right/w_LGL[-1]
    
    return rhs

# Scipy calculation
t_span = [0, 0.5]
t_eval=[0,0.1,0.2]
sol = solve_ivp(RHS_Neumann, 
                t_span=t_span, 
                y0=u0[1:], 
                method='RK45', 
                t_eval=t_eval, 
                args=(N, G, w_LGL, ux_right))

fig = plt.figure()
ax = fig.add_subplot(111)

Nt = len(t_eval)
for it in range(Nt):
    u = np.concatenate(( np.array([u_left]), 
                         sol.y[:,it] ))
    ax.plot(x, u, label='t={}'.format(t_eval[it]))


ax.legend()
ax.set_title(r'$u_t=u_{xx}$')
```

```python
# Interpolating to a finer grid for plotting purpose
y = np.linspace(-1,1,51)
T = InterpolatingMatrix(x, y, w_bary)

fig = plt.figure()
ax = fig.add_subplot(111)

Nt = len(t_eval)
for it in range(Nt):
    u = np.concatenate(( np.array([u_left]), sol.y[:,it] ))
    u_newgrid = np.dot(T, u)
    ax.plot(y, u_newgrid, label='t={}'.format(t_eval[it]))


ax.legend()
ax.set_title(r'$u_t=u_{xx}$')
```
