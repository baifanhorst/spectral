```python
# Testing the Fourier Differentiation Matrix
N = 40
J = np.arange(0,N)
x = 2 * np.pi / N * J
f = 1 / (3 + np.cos(x))
f_deri = np.sin(x) / (3 + np.cos(x))**2
D = CalcFourierDiffMat(N)
f_deri_DiffMat = np.dot(D, f)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, f_deri, label='Theoretical')
ax.scatter(x, f_deri_DiffMat, label='Diff Mat')
ax.legend()
ax.set_title(r'$f(x)=\frac{1}{3+\cos(x)}$')
```
