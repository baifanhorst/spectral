```python
# Testing the RHS 
cos = np.cos
sin = np.sin
pi = np.pi


N = 40
J = np.arange(0,N)
x = 2 * pi / N * J
f = 1 / (3+cos(x))
D = CalcFourierDiffMat(N)
nu = 0.01
RHS = CalcRHS_AdvDif_DiffMat(f, D, nu)
fx = sin(x) / (3+cos(x))**2
fxx = (cos(x) * (3+cos(x))**2 + sin(x)**2 * 2 * (3+cos(x))) / (3+cos(x))**4
RHS_th = nu * fxx - fx

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, RHS, label='Diff Mat')
ax.scatter(x, RHS_th, label='Theretical')
ax.legend()
ax.set_title(r'$f(x)=\frac{1}{3+\cos(x)}$')
```
