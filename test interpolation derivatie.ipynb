```python
import numpy as np
import matplotlib.pyplot as plt

# Reloading the module
import importlib

import OrthogonalPolynomials
importlib.reload(OrthogonalPolynomials)
from OrthogonalPolynomials import *
```

```python
# Testing the derivative calculation
N=10

s_nodes = LegendreGaussLobattoNodes(N)
w_bary = BarycentricWeights(s_nodes)
D = PolynomialDiffMatrix(s_nodes, w_bary)
t_nodes = np.pi/4 * (s_nodes+1)
y_nodes = np.sin(t_nodes)


s_nodes_plot = np.linspace(-1,1,101)
t_nodes_plot = np.pi/4 * (s_nodes_plot+1)


y_deri_th_plot = np.cos(t_nodes_plot)
y_deri_interp_plot = np.zeros(s_nodes_plot.shape)

y_interp_plot = np.zeros(s_nodes_plot.shape)
y_th_plot = np.sin(t_nodes_plot)

for i, s in enumerate(s_nodes_plot):
    y_deri_interp_plot[i] = LagrangeInterpolationDerivativeBary(s, s_nodes, 
                                                                y_nodes, w_bary, 
                                                                D)
    y_interp_plot[i] = LagrangeInterpolationBary(s, s_nodes, y_nodes, w_bary)
    
    
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(s_nodes_plot, y_deri_interp_plot, label='Interpolation')
ax.plot(s_nodes_plot, y_deri_th_plot, label='Theoretical')
ax.legend()

fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(s_nodes_plot, y_interp_plot, label='Interpolation')
ax.plot(s_nodes_plot, y_th_plot, label='Theoretical')
ax.legend()
```
