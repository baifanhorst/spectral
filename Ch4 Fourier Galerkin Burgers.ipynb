## $\nu=0$

```python
N = 100
nu = 0.0
f = lambda x: np.sin(x)
init_cond = CalcInitCond(f, N)
t_span = [0, 0.8]
t_eval = [0, 0.4, 0.8]
sol = solve_ivp(CalcRHS_Burgers, 
                t_span=t_span, 
                y0=init_cond, 
                method='RK45', 
                t_eval=t_eval, 
                args=(N, nu,))

N_eval = 50
x = np.linspace(0, 2*np.pi, N_eval)
NT = len(t_eval)
f_t = np.zeros((NT+1, N_eval))

for nt in range(NT):
    for j in range(N_eval):
        f_t[nt, j] = EvalSolution(x[j], sol.y[:,nt], N)

fig = plt.figure()
ax = fig.add_subplot(121)
for nt in range(NT):
    ax.plot(x, f_t[nt, :], label='t={}'.format(t_eval[nt]))

ax.legend()
ax.set_title(r"$u_0(x)=\sin(x)$")
```
